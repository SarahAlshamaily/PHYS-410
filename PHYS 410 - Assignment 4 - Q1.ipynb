```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
theta = np.linspace(0,100,1000)
r = (1/(np.sin(theta*3)))**(1/3)
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
plt.plot(x,y,label='streamlines')
plt.title("Streamlines")
plt.grid()
plt.legend()
plt.show()
```
